```python
from skimage import io
import matplotlib.pyplot as plt 
# im = io.imread('C:\\Users\\qSIM\\Pictures\\MOT_loading_fluo\\image_MOT11-8-20s.tif')
im = io.imread('loading_MOT.tif')
print(im.shape)
```

```python
offset=10
n_img=20
f = plt.figure(figsize=(30,3))
for j in range(n_img):
    ax = f.add_subplot(1,n_img,j+1)
    ax.imshow(im[j+offset],vmin=0,vmax=200)
```

```python
offset=10
n_img=14
roi=[27,47,32,49]                      #[xi,xf,yi,yf]
f = plt.figure(figsize=(30,3))
for j in range(n_img):
    ax = f.add_subplot(1,n_img,j+1)
    ax.imshow(im[2*j+offset][roi[0]:roi[1],roi[2]:roi[3]],vmin=0,vmax=250)
# plt.savefig('C:/Users/qSIM/Pictures/Presentation/MOT_loading.pdf')
```
